```python
def reduce_android_ver(version):
    if version.startswith('4'):
        return '4 and above'
    elif version.startswith(('1', '2', '3')):  # Check for versions starting with 1, 2, or 3
        return 'Under 4'
    else:
        return version  # Keep original value for other cases

df['Android Ver'] = df['Android Ver'].apply(reduce_android_ver)
```

```python
def reduce_android_ver(version):
    if version.startswith('4'):
        return '4 and above'
    elif version.startswith(('1', '2', '3')):  # Check for versions starting with 1, 2, or 3
        return 'Under 4'
    else:
        return version  # Keep original value for other cases

df['Current Ver'] = df['Current Ver'].apply(reduce_android_ver)
```

EDA - Exploratory Data Analysis


```python
df.groupby('Category')['Android Ver'].sum().sort_values(ascending= False).head(10)
```

```python
df.groupby('Category')['Rating'].sum().sort_values(ascending= False).head(10).plot(kind = 'bar')
plt.xlabel('App')
plt.ylabel('Total Rating')
plt.title('Top 10 Apps by Total Rating')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
df.groupby('Category')['Reviews'].sum().sort_values(ascending= False).head(10).plot(kind = 'bar')
plt.xlabel('App')
plt.ylabel('Total Rating')
plt.title('Top 10 Apps by Total Rating')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
